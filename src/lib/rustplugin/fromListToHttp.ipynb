```python
import json
import http.client
import requests
import asyncio
import time
```

```python
file_name = "./mystarred.json"
data = ''
with open(file_name) as json_file:
    data = json.load(json_file)
    for p in data:
        print('Name: ' + p)
```

```python
print(data[0:3])
```

```python
to_do_list = data[0:3]
```

```python
# slice [18:]
```

```python
for task in to_do_list:
    print(task[18:])  
    location = task[18:]
    URL = task[:18]
    PARAMS = {'address':location} 
    r = requests.get(url = URL, params = PARAMS)
    f = open("test.txt", "a")
    f.write("========================" + r.text)
    f.close()
    await asyncio.sleep(5)
```
